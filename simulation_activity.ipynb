```python
dice1 = np.array([1, 2, 3, 4, 5, 6, 5, 4, 3, 2])
dice2 = np.array([1, 1, 2, 2, 4, 5, 6, 3, 2, 1])
```

```python
sum_rolls = np.add(dice1, dice2)
print(sum_rolls)
```

```python
print("Average of dice1:", np.mean(dice1))
print("Average of dice2:", np.mean(dice2))
```

```python
print("Median of dice1:", np.median(dice1))
print("Median of dice2:", np.median(dice2))
```

```python
product_rolls = np.multiply(dice1, dice2)
print(product_rolls)
```

```python
two_dice = np.array([dice1, dice2])
print(np.mean(two_dice, axis=1))
```

```python
unique, counts = np.unique(dice1, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
import matplotlib.pyplot as plt

unique, counts = np.unique(dice1, return_counts=True)

plt.bar(unique, counts)
plt.xlabel("Dice Value")
plt.ylabel("Frequency")
plt.title("Frequency of Dice1")
plt.show()
```
