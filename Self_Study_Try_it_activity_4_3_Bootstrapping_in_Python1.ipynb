# Self-Study Try-it Activity 4.3 – Bootstrapping in Python (Completed)

```python
import numpy as np
```

## Original Data

```python
data = np.array([10, 12, 13, 12, 11, 14, 13, 15, 16, 12])
print(data)
```

```python
import numpy as np

data = np.array([10, 12, 13, 12, 11, 14, 13, 15, 16, 12])
```

## Bootstrap Sampling

```python
n_iterations = 1000
sample_size = len(data)

means = []

for _ in range(n_iterations):
    sample = np.random.choice(data, size=sample_size, replace=True)
    means.append(np.mean(sample))

print("First 10 bootstrap means:", means[:10])
```

## Estimate Mean and Confidence Interval

```python
bootstrap_mean = np.mean(means)
lower = np.percentile(means, 2.5)
upper = np.percentile(means, 97.5)

print('Bootstrap mean:', bootstrap_mean)
print('95% Confidence Interval:', lower, 'to', upper)
```

## Visualization

```python
import matplotlib.pyplot as plt

plt.hist(means, bins=30)
plt.title('Bootstrap Distribution of Mean')
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.show()
```
