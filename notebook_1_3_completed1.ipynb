## Dice Simulation

```python
dice = np.random.randint(1, 7, 1000)

print("Average dice value:", np.mean(dice))
print("Counts:", np.unique(dice, return_counts=True))
```

## Probability Estimation

```python
prob_6 = np.sum(dice == 6) / len(dice)
print("Probability of rolling 6:", prob_6)
```

## Law of Large Numbers

```python
import numpy as np

means = []

for i in range(1, 1001):
    sample = np.random.randint(1, 7, i)   # ← MUST be indented
    means.append(np.mean(sample))

print("Last mean:", means[-1])
```
